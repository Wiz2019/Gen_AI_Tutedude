TASK 3: Product Pricing (Dictionaries)

```python
#1 create a dictionary price named price_dict with product names as keys and their prices as values
price_dict = {
    "Laptop": 999.99,
    "Mouse": 25.99,
    "Keyboard": 45.99,
    "Monitor": 199.99,
    "Desk": 150.00,
    "Chair": 85.00
}
print("Initial price dictionary:", price_dict)

#2 add a new product and its price to the dictionary
price_dict["Headphones"] = 79.99
#2a update the price of an existing product
price_dict["Laptop"] = 899.99
#2b remove a product from the dictionary(handle the case where the product might not exist)
if "Monitor" in price_dict:
    del price_dict["Monitor"]
    print("Monitor removed from price_dict.")
else:
    print("Monitor not found in price_dict.")
print("Updated price dictionary:", price_dict)

#3 print the average price of all products in the dictionary
average_price = sum(price_dict.values()) / len(price_dict)
print("Average price of products:", average_price)

#4 Extra (optional) - print the product with both the highest and lowest price
max_price_product = max(price_dict, key=price_dict.get)
min_price_product = min(price_dict, key=price_dict.get)
print("Product with the highest price:", max_price_product, "-", price_dict[max_price_product])
print("Product with the lowest price:", min_price_product, "-", price_dict[min_price_product])
```

TASK 4: Combined Operations

```python
#1 using the products list and price_dict create a new list of tuples names catalog where each tuple is (product_name, price, category)
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Desk", "Chair", "Headphones"]
categories = ["Electronics", "Accessories", "Accessories", "Electronics", "Furniture", "Furniture", "Accessories"]
price_dict = {
    "Laptop": 999.99,
    "Mouse": 25.99,
    "Keyboard": 45.99,
    "Monitor": 199.99,
    "Desk": 150.00,
    "Chair": 85.00,
    "Headphones": 79.99
}
catalog = []
for i in range(len(products)):
    product = products[i]
    price = price_dict[product]
    category = categories[i]
    catalog.append((product, price, category))
print("Catalog of products:", catalog)

#2 from catalog create a new dictionary category_to_products that maps each category to a list of products in that category
category_to_products = {}
for product, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    category_to_products[category].append(product)
print("Category to products mapping:", category_to_products)

#3 print all products that belong to the category that has the maximim number of products
max_category = max(category_to_products, key=lambda c: len(category_to_products[c]))
print("Category with maximum products:", max_category)
print("Products in that category:", category_to_products[max_category])
```
