Task1 Basic Function : Price after discount 

```python
#1 write a function apply_discount(price, discount_percentage) that takes a price and a discount percentage and returns the discounted price.

def apply_discount(price, discount_percentage = 5):
    if discount_percentage > 60:
        print("Discount percentage cannot be more than 60%. Setting it to 60%.")
        discount_percentage = 60
    discount_amount = price * (discount_percentage / 100)
    discounted_price = price - discount_amount
    return discounted_price
```

Task 2 Recursive Function : Factorial Utility 

```python
# Create a recursive funtion factorial(n)

def factorial(n):
    if n < 0:
        print("Factorial is not defined for negative numbers.")
        return None
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

TASK 5 Using filter(): Filter Expensive Products

```python
# Given a list prices use filter() to create a list with prices > 500 and one <= 500 and print both lists

prices = [100, 250, 400, 1200, 50, 2000, 850]
prices_above_500 = list(filter(lambda price: price > 500, prices))
prices_below_500 = list(filter(lambda price: price <= 500, prices))
print("Prices above 500:", prices_above_500)
print("Prices 500 or below:", prices_below_500)
```

TASK 6 Combined Utility Function

```python
# write a function process_price(prices) which creates a list with map and compute 10% and another list which filters the discounted price > 300

def process_price(prices):
    discounted_prices = list(map(lambda price : apply_discount(price, 10), prices))
    filtered_prices = list(filter(lambda price : price > 300, discounted_prices))
    return discounted_prices, filtered_prices

process_price([100, 500, 900, 50, 750])
```

TASK 7 Mini Problem : Menu Using Functions

```python
# create three functions add_price(price_list, price), get_average_price(price_list), get_max_price(price_list) 
# then create a simple menu using loops

def add_price(price_list, price):
    price_list.append(price)
    print(f"Price {price} added. Current price list: {price_list}")

def get_average_price(price_list):
    if len(price_list) == 0:
        print("Price list is empty. Cannot compute average.")
        return None
    average = sum(price_list) / len(price_list)
    print(f"Average price: {average}")
    return average

def get_max_price(price_list):
    if len(price_list) == 0:
        print("Price list is empty. Cannot compute maximum price.")
        return None
    max_price = max(price_list)
    print(f"Maximum price: {max_price}")
    return max_price

prices = []
while True:
    print("\nMenu:")
    print("1. Add a price")
    print("2. Get average price")
    print("3. Get maximum price")
    print("q. Quit")
    
    user_input = input("Enter your choice: ")
    
    if user_input == '1':
        price_input = input("Enter the price to add: ")
        # Simple input validation - check if input is a valid number
        if price_input and all(c.isdigit() or c == '.' for c in price_input) and price_input.count('.') <= 1:
            price = float(price_input)
            add_price(prices, price)
        else:
            print("Invalid input. Please enter a numeric value for the price.")
            continue
    elif user_input == '2':
        get_average_price(prices)
    elif user_input == '3':
        get_max_price(prices)
    elif user_input == 'q':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please select a valid option from the menu.")
        continue
```
