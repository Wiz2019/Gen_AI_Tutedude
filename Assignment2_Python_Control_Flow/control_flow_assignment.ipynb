Task 1 Discount Rules (if/ elif/ else)

```python
#1 WAP that read integer from the user using the variable order_amount
order_amount_str = input("Enter the order amount: ")

#2&3 Apply the discount rules and also check if the input is not a non-numeric value by displaying an error message
try:
    order_amount = int(order_amount_str)
except ValueError:
    print("Invalid input. Please enter a numeric value for the order amount.")
else:
    if order_amount >= 2000:
        discount = order_amount * 0.15
    elif 1500 <= order_amount < 2000:
        discount = order_amount * 0.10
    elif 1000 <= order_amount < 1500:
        discount = order_amount * 0.07
    else:
        discount = 0

    final_amount = order_amount - discount
    print(f"The final amount after discount is: {final_amount}")

    # Extra(Optional) add tax(5%) to the final amount after discount and display subtotal, tax and final total
    tax = final_amount * 0.05
    print(f"The subtotal after discount is: {final_amount}")
    print(f"The tax amount is: {tax}")
    print(f"The final total after adding tax is: {final_amount + tax}")
```

TASK 2 Process Multiple Orders (for loop)

```python
#1 given a list orders apply the discount rules and print order amount, discount % and final amount

orders = [1200, 2500, 800, 1750, 3000]
discounted_amount = []
for order in orders:
    if order >= 2000:
        discount = order * 0.15
        discount_percent = 15
    elif 1500 <= order < 2000:
        discount = order * 0.10
        discount_percent = 10
    elif 1000 <= order < 1500:
        discount = order * 0.07
        discount_percent = 7
    else:
        discount = 0
        discount_percent = 0

    final_amount = order - discount
    discounted_amount.append(final_amount)
    print(f"Order Amount: {order}, Discount: {discount_percent}%, Final Amount: {final_amount}")

#2 Compute and print the total revenue after discounts

total_revenue = sum(discounted_amount)
print(f"Total Revenue After Discounts: {total_revenue}")

# Extra(Optional) Print the number of orders that recieved a discount 
discounted_orders_count = sum(1 for order in orders if order >= 1000)
print(f"Number of Orders that Received a Discount: {discounted_orders_count}")
```

TASK 3 User Menu (While loop + break/continue)

```python
#1&2 Use a while loop to ask the user for options and q to exit it and use continue for invalid options and break when q is entered.

while True:
    user_input = input("Enter options as '1' or '2' or 'q' to exit: ")
    if user_input == '1':
        new_order_str = input("Enter the new order amount: ")
        try:
            new_order = int(new_order_str)
        except ValueError:
            print("Invalid amount; must be an integer.")
            continue
        orders.append(new_order)
        print(f"New order added. Current Orders: {orders}")
    elif user_input == '2':
        total = 0
        for o in orders:
            if o >= 2000:
                discount = o * 0.15
            elif 1500 <= o < 2000:
                discount = o * 0.10
            elif 1000 <= o < 1500:
                discount = o * 0.07
            else:
                discount = 0
            final = o - discount
            total += final
        print(f"All Orders: {orders}")
        print(f"Total revenue after discounts: {total}")
    elif user_input == 'q':
        print("Exiting the program.")
        break
    else:
        print("Invalid option. Please enter '1', '2', or 'q'.")
        continue
```

TASK 5 Loop Control with Conditions (break & Continue)

```python
#1 Given a daily sales list iterate through it and apply given logic print the final total
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0
for sale in daily:
    if sale == -1:
        print("Corrupted data found. Stopping the process.")
        break
    elif sale == 0:
        print("No sales for the day.")
        continue
    else:
        total_sales += sale
        print(f"Total Sales: {total_sales}")

#2 Print the final total after the loop completes (or stops due to corrupted data)    
print(f"Total Sales after the complete loop: {total_sales}")
```
