TASK 1 Write Sales Records to a file

```python
# write the list to a file named sales.txt in write mode write sales in each new line close the file and then reopen and print its content

sales = [1200, 450, 980, 1500, 3000]

# Write sales to file in write mode
file = open('sales_data.txt', 'w')
for sale in sales:
    file.write(str(sale) + '\n')
file.close()

# Reopen file in read mode and print its content
with open('sales_data.txt', 'r') as file:
    content = file.read()
    print(content)


#optional write the data in comma-separated format instead of new line separated format

with open('sales_comma.txt', 'w') as file:
    file.write(','.join(str(sale) for sale in sales))

# Read and print the comma-separated file
with open('sales_comma.txt', 'r') as file:
    content = file.read()
    print("Comma-separated format:")
    print(content)
```

TASK 2 Read Files In Different Ways 

```python
# open sales.txt read the entire file with read() and print it and also read the first line with readline() and read all lins using .readlines() and 
# convert them into a list of integers and ensure cleanup of newline characters

with open('sales_data.txt', 'r') as file:
    # Read entire content
    content = file.read()
    print("Entire file content:")
    print(content)

    # Read first line
    file.seek(0)
    first_line = file.readline().strip()
    print("First line:")
    print(first_line + '\n')

    # Read all lines and convert to list of integers
    file.seek(0)
    lines = file.readlines()
    sales_list = [int(line.strip()) for line in lines]
    print("List of sales as integers:")
    print(sales_list)

```

TASK 3 Append New Sales

```python
# Append the new sales and then reopen and print its updated content

with open('sales_data.txt', 'a') as file:
    new_sales = [5000, 2500, 1700]
    for sale in new_sales:
        file.write(str(sale) + '\n')

# Reopen and print updated content
with open('sales_data.txt', 'r') as file:
    updated_content = file.read()
    print("Updated file content:")
    print(updated_content)

# Optional: Print the total number of lines in sales.txt

with open('sales_data.txt', 'r') as file:
    number_of_lines = len(file.readlines())
    print(f"Total number of lines in sales_data.txt: {number_of_lines}")
```

TASK 4 Generate Summary Report From File

```python
# Read the sales_data file and convert them to int calculate and print total sales, highest sale, Lowest sale, Average Sale

with open('sales_data.txt', 'r') as file:
     lines = file.readlines()
     sales_summary_list = [int(line.strip()) for line in lines]
     total_sales = sum(sales_summary_list)
     highest_sale = max(sales_summary_list)
     lowest_sale = min(sales_summary_list)
     average_sale = total_sales/len(sales_summary_list)
     print(f"Total Sales: {total_sales}")
     print(f"Highest Sale: {highest_sale}")
     print(f"Lowest Sale: {lowest_sale}")
     print(f"Average Sale: {average_sale}")
```

TASK 5 Create Product info File (User Input)

```python
# Ask the user for 3 produuct name and their price and write them in a file named products.txt in this format ProductName | Price and then read the 
# file and print each line

with open('products.txt', 'w') as file:
    for i in range(3):
        product_name = input("Enter product name: ")
        price = input("Enter product price:")
        file.write(f"{product_name} | {price}\n")

# Read and print the products file
with open('products.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

TASK 6 Read file Safely (Error Handling Inside File Handling Only)

```python
# Ask the user for a file name to open and if it exists read and print it else if the it does not exits print "File not found ..." 

import os

file_name = input("Enter the file name to print:")
if os.path.exists(file_name):
    with open(file_name, 'r') as file:
        content = file.read()
        print(content)
else:
    print("File not found. Please check the filename.")
```

TASK 7 Mini Project - Export Discounted Prices

```python
# create a dictonary and ask the user for discount % and write the discounted price in a file named discount_report.txt as 
# ProductName | Original Price | Discounted Price and read it and print it

prices = { "Mouse" : 500, "Keyboard" : 800, "Monitor" : 7000, "Pendrive" : 400, "camera" : 5000}

discount_percentage = float(input("Enter discount percentage: "))
with open('discount_report.txt', 'w') as file:
    for product, price in prices.items():
        discounted_price = price - (price * discount_percentage / 100)
        file.write(f"{product} | {price} | {discounted_price}\n")

# Optional Write a summary total items and Average discounted price at the end of the file and print it

with open('discount_report.txt', 'a') as file:
    total_items = len(prices)
    average_disounted_price = sum(price - (price * discount_percentage / 100) for price in prices.values()) / total_items
    file.write(f"\nTotal Items: {total_items} | Average Discounted Price: {average_disounted_price}\n")

# Read and print the updated discount report
with open('discount_report.txt', 'r') as file:
    for line in file:
        print(line.strip()) 
```
